## Le calcul symbolique (littéral) avec le module sympy
On peut faire du calcul littéral avec sympy.
   * On a l'expression $x+2y$.

   * Multiplier cette expression par $x$

   * Puis developper cette dernière expression.
   

Vérifier votre résultat avec executant la cellule ci dessous avec **SHIFT+Entrée**

````python
x, y = symbols('x y')


expression = x+2*y
print(expression)
print('Multiplions cette expression par x')

expression2 = x*expression
print("On a alors l'expression:")
pprint(expression2)
print()
print('écriture développée:')
pprint(expand(expression2))
````
