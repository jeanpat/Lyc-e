## Comment les statistiques peuvent aider à nettoyer une image?
Chargeons trois images

```python
lena = nd.imread('LenaBW.png')
saltPepper = nd.imread('LenaSaltPepper.png')
noisy = nd.imread('Lena_noisy.png')
```

Afffichons les:

```python
plt.figure(figsize=(12,4))
plt.subplot(131)
plt.title("lena")
plt.imshow(lena,cmap="gray")
plt.subplot(132)
plt.title("lena poivre et sel")
plt.imshow(saltPepper,cmap="gray")
plt.subplot(133)
plt.title("lena bruitée")
plt.imshow(noisy,cmap="gray")
plt.show()
```

En zoomant sur les images, comment justifier le titre de l'image centrale?

```python
zoom_sp = saltPepper[250:292,240:285]
zoom2_sp = saltPepper[276:281,274:279]
array_sp = zoom2_sp.flatten()

zoom = lena[250:292,240:285]
zoom2 = lena[276:281,274:279]
array_lena = zoom2.flatten()

zoom_ns = noisy[250:292,240:285]
zoom2_ns = noisy[276:281,274:279]
array_ns = zoom2_ns.flatten()
```

```python
plt.figure(figsize=(12,4))

plt.subplot(231)
plt.imshow(zoom,cmap="gray", vmin=0, vmax=255)
plt.subplot(232)
plt.imshow(zoom_sp,cmap="gray", vmin=0, vmax=255)
plt.subplot(233)
plt.imshow(zoom_ns,cmap="gray", vmin=0, vmax=255)
plt.show()
plt.subplot(234)
plt.imshow(zoom2,cmap="gray", vmin=0, vmax=255)
plt.subplot(235)
plt.imshow(zoom2_sp,cmap="gray", vmin=0, vmax=255)
plt.subplot(236)
plt.imshow(zoom2_ns,cmap="gray", vmin=0, vmax=255)
plt.show()
```

## Ce code permet de filtrer une image

```python
lena_median = nd.median_filter(lena, size=(5,5)) 
```

```python
plt.subplot(131)
plt.title("lena originale")
plt.imshow(lena_median, cmap="gray")

plt.subplot(132)
plt.title("lena poivre sel")
plt.imshow(saltPepper, cmap="gray")

plt.subplot(133)
plt.title("lena bruitée filtre median")
plt.imshow(lena_median, cmap="gray")
plt.show()
```
