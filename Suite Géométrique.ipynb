   * accrocher 4 à la variable $x$

      ```Python
      x = 4
      x+=1
      ```
      
## Coefficient multiplicateur

On choisit une valeur initiale, $x=4$, que l'on souhaite augmenter de **50%**.
Cette valeur, sera la nouvelle valeur de $x$.

Commençons par multiplier $x$ par $1.50$, c'est le coefficient multiplicateur associé à une augmentation de 50%

$$4 \times 1.50=6$$

cette multiplication  donne le même résultat que:

$$4 + \frac{50}{100}\times4= 4+2=6$$

### Coefficient multiplicateur associé à une augmentation de p%: 

Pour augmenter une quantité de $p\%$, il faut la multiplier par un **cm** égal à:

$$cm = \dfrac{100+p}{100}$$

### Coefficient multiplicateur associé à une diminution  de p%: 

Pour diminuer une quantité de $p\%$, il faut la multiplier par un **cm** égal à:

$$cm = \dfrac{100-p}{100}$$

### Exemple de calcul en python

Choisissons une valeur initiale que l'on va augmenter de 50%, pour cela on va écrire:

```Python
u = 200
print("La valeur initiale de u est:", u)
u = 1.5*u
print("La nouvelle valeur de u est:", u)
```

# Suite géométrique:
Une suite géométrique est une suite de nombres définie par un premier terme $U_0$ ou $U_1$ (le plus souvent) et une raison $q$


$$
\left( U_n \right) = \left\{
    \begin{array}{ll}
        U_1 \\
        U_{n+1}=q \times U_{n}
    \end{array}
\right.
$$
avec:
$$
\begin{array}{ll}
        U_1=4 \\
        q=1.5
    \end{array}
$$

Comment calculer le terme suivant avec python ? On va faire les choses suivantes:

   * Dans un premier temps on **affecte** la valeur 4 à la variable $U$ ou encore on attache l'étiquette $U$ au nombre 4.

   * **=** signifie *affecter*
   
   * Ensuite on a itéré (répété) la réaffectation de la variable $U$ en utilisant l'ancienne valeur de $U$ augmentée de 50%.


## Sens de variation d'une suite géométrique
De manière générale:
   * Une suite $\left( U_n \right)$ est strictement croissante si:

$$U_{n+1}>U_n$$
   * Une suite $\left( U_n \right)$ est strictement décroissante si:

$$U_{n+1}<U_n$$

Une **suite géométrique** de premier terme strictement **positif** est:
   * croissante si $q>1$
   * décroissante si $q <1$
