## Encadrement de $\sqrt{2}$
On a donc encadré la longueur du coté du carré d'aire égale à 2:
$$1,69 < 2< 2.25$$
c'est à dire:
$$ a^2 < 2 <b^2$$
On peut donc encadrer la longueur du côté du carré:
$$1.3<côté<1.5$$

Comme $a=1.3$ et $b=1.5$, on a encadré la longueur du coté du carré violet. La longueur du coté de ce carré vaut:

$$\Huge{\color{blue}{côté}= \sqrt{2}}$$



encadrement.png

On a donc :
$$ 1,3 < \sqrt{2}< 1.5 $$
ou encore:
$$ a < \sqrt{2}< b $$


### Rechercher à la main d'un encadrement de $\sqrt{2}$:

1. Rechercher un valeur de $a$ plus proche de $\sqrt{2}$:

    * Choisir une valeur de $a> 1.3$
    * Cette valeur doit être telle que: $a^2 < 2$
2. Rechercher un valeur de $b$ plus proche de $\sqrt{2}$:

    * Choisir une valeur de $b <1.5$
    * Cette valeur doit être telle que:  $b^2 > 2$

Dans la cellule ci dessous, modifier la valeur de $a$ qui respecte les deux conditions:

````python
# Modifier la valeur de a
a = Fraction("1.4")

## Ne pas modifier ci-dessous
aire = Fraction("2")
racine = u"\u221a" +"2"

if a > Fraction("1.3"):
  print("a=",a,"~",float(a)," donc la première condition a>1.3  est respectée:")
  if a*a < aire :
    print("La deuxième condition est respectée, a²=", a*a, "~",float(a*a)," est bien inférieur à 2")
    print("       donc ",float(a), " < "+racine)
  else:
    print("La deuxième condtion n'est pas respectée, a²=", a*a, "~",float(a*a)," est supérieur à 2")
    print("       donc ",float(a), " > "+racine)
else:
  print("a =", a, "~",float(a)," est trop petit")
````
